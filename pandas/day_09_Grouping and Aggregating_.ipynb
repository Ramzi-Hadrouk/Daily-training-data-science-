```python
import pandas as pd
```

```python
# Create a sample series
numbers = pd.Series([1, 2, 3,4, 4, 100,100])

```

```python
print( numbers.mean() )   
```

```python
print(numbers.median() ) 
```

```python
print(numbers.mode() ) 
```

```python
print(numbers.count() ) 
```

```python
print(numbers.value_counts() ) 
```

```python
print(numbers.value_counts(normalize=True) ) 
```
