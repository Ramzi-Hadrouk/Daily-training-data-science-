```python
import pandas as pd
```

```python
df=pd.read_csv("/home/geek/Desktop/DE/data/world_population.csv")
pd.set_option('display.max_rows', 30)
df[['Country','Continent','2022 Population']]
```

```python
df.sort_values(by=['2022 Population','Country'] ,ascending=[True,False],inplace=True )
df[['Country','Continent','2022 Population']]
```

```python
df.sort_index()
```

```python
df.nlargest(columns=['2022 Population'] ,n=10)
```

```python
df.nsmallest(columns=['2022 Population'] ,n=10)
```
