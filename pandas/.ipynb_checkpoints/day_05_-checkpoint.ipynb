```python
 import pandas as pd
```

```python
people = {
    "first": ["Corey","Jane", "John", "Emily", "Michael", "Sarah", "David", "Emma", "Chris", "Olivia"],
    "last": ["Schafer", "Doe", "Doe", "Smith", "Brown", "Johnson", "Miller", "Wilson", "Moore", "Taylor"],
    "email": [
        "CoreyMSchafer@gmail.com",
        "JaneDoe@email.com",
        "JohnDoe@email.com",
        "EmilySmith@email.com",
        "MichaelBrown@email.com",
        "SarahJohnson@email.com",
        "DavidMiller@email.com",
        "EmmaWilson@email.com",
        "ChrisMoore@email.com",
        "OliviaTaylor@email.com"
    ]
}
```

```python
df=pd.DataFrame(people)
```

```python
lista= [df['first'].str.lower().str.startswith('e')]

print(lista[0][3])
lista
```

```python
df[df['first'].str.lower().str.startswith('e')]
```

```python
filter_one=df['first' ].str.startswith('E') #more readable way 
df.loc[filter_one]
```

```python
df.loc[filter_one, 'email']
```

```python
filter_one= ( df['first' ].str.startswith('E')) | (df['last']=='Doe' ) #more readable way 
filter_one
```

```python
df.loc[~filter_one]
```
